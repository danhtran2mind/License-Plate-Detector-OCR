# Evaluation

```python
# Evaluation
from ultralytics import YOLO

# Load the model
model = YOLO('./ckpts/finetune/runs/license_plate_detector/weights/best.pt')

# Evaluate on the test set
results = model.val(data='/kaggle/input/license-plate-detection/data.yaml',
                    split='test', imgsz=640, batch=64, device='0',
                    save=True, save_txt=True, save_conf=True,
                    project="./ckpts/finetune/runs",
                    name="license_plate_detector-test"
                    )
```
